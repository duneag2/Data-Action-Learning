```python
import pandas as pd
import numpy as np
```

```python
g = pd.read_csv('./gocd.csv')
z = pd.read_csv('./zzang9.csv')
```

```python
g_content = g['content'].to_list()
z_content = z['content'].to_list()
```

```python
g_re = [x[2:-2] for x in g_content]
z_re = [x[2:-2] for x in z_content]
```

```python
g_re
```

```python
g_result = []

# gocd 삭제
for content in g_re :
    string = content.replace("모든 게시물은 사진 + 글 첨부 필수","")
    string = content.replace("질문글 삭제시 활동정지","")
    string = content.replace("... 더보기","")
    string = content.replace("&","")
    g_result.append(string)

g['content'] = g_result
```

```python
g
```

```python
z_result = []

# zzang9 삭제
for content in z_re :
    string = content.replace("짱구대디랑 카페 체험단은 브랜드로부터 진행 비용 및 제품을 제공 받아 진행하는 이벤트입니다","")
    string = string.replace("패션질문 게시판 이용시, 관련사진 첨부 의무","")
    string = string.replace("※글 작성시 위 배너를 삭제하시면 활동 제제 & 글 무통보 삭제 대상이오니 유의하시기 바랍니다","")
    string = string.replace("해당게시판은 생방송 컨텐츠에 사용 될 수 있습니다","")
    string = string.replace("착장 기본 양식을 지켜주세요.","")
    string = string.replace("데일리룩 관련 공지","")
    string = string.replace("<기본양식>","")
    string = string.replace("*쇼핑후기 게시판은 제품 실사 혹은 실사용 후기가 필수입니다. 누락 될 경우 삭제 될 수 있으니 참고해주세요.","")
    string = string.replace("🔈","")
    
    z_result.append(string)
    

z['content'] = z_result
z
```

```python
# g.to_csv('./gocd_p.csv',encoding='utf-8-sig',index = False)
# z.to_csv('./zzang9_p.csv',encoding='utf-8-sig',index = False)
result = pd.concat([g,z])
result.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
result.to_csv('./delete_result.csv',encoding='utf-8-sig',index=False)
```
