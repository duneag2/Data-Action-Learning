```python
from konlpy.tag import Okt
from collections import Counter
```

```python
new_data = pd.DataFrame(df1, columns = ['title', 'content'])
```

```python
new_data
```

```python
text_data=[]
nlp = Okt()
for i in range(len(new_data.index)):
    row_text = nlp.nouns(new_data['content'][i])
    for j in range(len(row_text)):
        if len(row_text[j])>=2:
            text_data.append(row_text[j])
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = "모든 게시 사진 첨부 질문 삭제 활동 정지 보기 생각 필수 구매 카페 제품"

result = []
for w in text_data:
    if w not in stop_words:
        result.append(w)
```

```python
result
```

```python
word = dict(Counter(result).most_common(50))
word
```
