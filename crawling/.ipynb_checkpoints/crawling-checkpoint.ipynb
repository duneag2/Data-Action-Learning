```python
# ì¹´í˜ ì •ë³´
gocd = 'https://cafe.naver.com/casuallydressed'
zzang9 = 'https://cafe.naver.com/zzang9daddy'
        
cafe_url = zzang9
cafe_name = 'zzang9'
keywords = ['êµ¬ì œ']


# í¬ë¡¬ ë“œë¼ì´ë²„ ì‹¤í–‰
driver = webdriver.Chrome('C:\chromedriver.exe')
driver.get("https://nid.naver.com/nidlogin.login")

# ë¡œê·¸ì¸ ì •ë³´
login = {"id" : ""   # ë„¤ì´ë²„ ì•„ì´ë””
        ,"pw" : ""   # ë„¤ì´ë²„ ë¹„ë°€ë²ˆí˜¸
        }

# ë¡œê·¸ì¸ ì •ë³´ ì…ë ¥ í•¨ìˆ˜
def clipboard_input(user_xpath, user_input):
    temp_user_input = pyperclip.paste()

    pyperclip.copy(user_input)
    driver.find_element_by_xpath(user_xpath).click()
    ActionChains(driver).key_down(Keys.CONTROL).send_keys('v').key_up(Keys.CONTROL).perform()

    pyperclip.copy(temp_user_input)
    time.sleep(1)
    
# id, pw ì…ë ¥ í›„ í´ë¦­
clipboard_input('//*[@id="id"]', login.get("id"))
clipboard_input('//*[@id="pw"]', login.get("pw"))
driver.find_element_by_xpath('//*[@id="log.login"]').click()

time.sleep(15)

# ì¹´í˜ ê²Œì‹œê¸€ ë§í¬ ìˆ˜ì§‘

for keyword in keywords:
    ### ì¹´í˜ ì£¼ì†Œ ì…ë ¥
    driver.get(cafe_url)
    
    ### í‚¤ì›Œë“œ ê²€ìƒ‰
    clipboard_input('//*[@id="topLayerQueryInput"]', keyword)
    try :
        driver.find_element_by_xpath('//*[@id="cafe-search"]/form/button').click()   # ì™¼ìª½ì— ê²€ìƒ‰ì°½
    except:
        driver.find_element_by_xpath('//*[@id="info-search"]/form/button').click()   # ì˜¤ë¥¸ìª½ì— ê²€ìƒ‰ì°½
    driver.implicitly_wait(0.5)
    driver.switch_to.frame('cafe_main')

    ### í‚¤ì›Œë“œ ìˆ˜ì§‘ ì •ë³´
    num_per_page = 15
    address_list=[]
    page = 1
    
    l=True
    while l:
        time.sleep(random.randint(0,5) )
        
        ### í˜„ì¬ í˜ì´ì§€ì˜ html ë¶ˆëŸ¬ì˜¤ê¸°
        r = driver.page_source
        page_html = BeautifulSoup(r, "html.parser")
        content = page_html.find("div", class_="article-board result-board m-tcol-c").find('tbody')
        body = content.find_all("tr")

        ### ê²Œì‹œê¸€ ì •ë³´ ì €ì¥í•˜ê¸°
        for x in body:
            temp_dict={}
            if x.find("div", class_="board-number") is not None:
                temp_dict['no'] = x.find("div", class_="board-number").text.strip()
                temp_dict['title'] = x.find("div", class_="board-list").text.strip().replace('  ','').replace('\n','')
                temp_dict['link'] = x.find('a').get('href')
                temp_dict['name'] = x.find("td", class_="td_name").find('a',class_='m-tcol-c').text.strip()
                temp_dict['date'] = x.find("td", class_="td_date").text.strip()
                temp_dict['view'] = x.find("td", class_="td_view").text.strip()
                address_list.append(temp_dict)
        print(str(page) +"page done")

        ### ë‹¤ìŒ í˜ì´ì§€ë¡œ ë„˜ì–´ê°€ê¸°
        page+=1
        driver.implicitly_wait(1)
        try:
            if page<=10:   # 1~10 : í˜ì´ì§€ ë²ˆí˜¸ ê·¸ëŒ€ë¡œ
                page_xpath = str(page)
                driver.find_element_by_xpath('//*[@id="main-area"]/div[7]/a[' + page_xpath + ']').click()
            elif page == 11:   # 11 : ë‹¤ìŒ ë²„íŠ¼
                driver.find_element_by_xpath('//*[@id="main-area"]/div[7]/a[11]/span').click()
            elif page>11 and page%10!=1:   # 12~ : í˜ì´ì§€ ë²ˆí˜¸ ë§ˆì§€ë§‰ ìë¦¬ + 1
                page_xpath = str(page-((page-1)//10)*10+1)
                driver.find_element_by_xpath('//*[@id="main-area"]/div[7]/a[' + page_xpath + ']').click()
            elif page%10 == 1:   # 21,31.. : ë‹¤ìŒ ë²„íŠ¼
                driver.find_element_by_xpath('//*[@id="main-area"]/div[7]/a[12]/span').click()
        except:
                address_df = pd.DataFrame(address_list)
                address_df['idx_no'] = range(1,len(address_df)+1)   # ì¡°ì¸í•  í‚¤ ê°’
                address_df.to_pickle(save_path+"cafe_address_"+cafe.get("name")+"_"+keyword+".pkl")
                print("(í˜„ì¬ì‹œê°) "+str(datetime.datetime.now())+": done")
                l=False
                
        if page > 30 : # í˜ì´ì§€ ìµœëŒ€ 30ê°œ í¬ë¡¤ë§
            break
            
# ì œëª©, ë§í¬, ì‘ì„±ì, ë‚ ì§œ, ì¡°íšŒìˆ˜ ë‹´ê¸´ ë°ì´í„°í”„ë ˆì„ ìƒì„±       
address_df = pd.DataFrame(address_list)
ad = address_df['link']
all_address = "https://cafe.naver.com" + ad

# ì œëª© ë‚´ìš© í¬ë¡¤ë§
all_content = []
all_title = []

i = 1 # for count content
for url in all_address:
    try:
        # í˜ì´ì§€ ì´ë™
        driver.get(url)
        try:
            driver.switch_to.frame("cafe_main")
            driver.sleep(1)
        except:
            pass

        # ì œëª© í¬ë¡¤ë§
        title = driver.find_element_by_css_selector('#app > div > div > div > div > div > div > h3').text
        all_title.append(title)

        # ë‚´ìš© í¬ë¡¤ë§
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        content = soup.select('div.article_viewer')

        cur_content = []
        for c in content:
            cur_content.append(re.sub('\xa0', ' ', c.text.strip()))
        all_content.append(cur_content)
        if i % 10 == 0:
            print(str(i)+" content done")
        i+=1

    except:
        pass    

# í¬ë¡¤ë§ ê²°ê³¼ csv íŒŒì¼ë¡œ ì €ì¥
# result_df = pd.DataFrame.from_dict({'title': all_title, 'content': all_content}, orient='index')   
# result_df = result_df.transpose()
# result_df.to_csv('./'+cafe_name +'.csv',encoding='utf-8-sig')
```
